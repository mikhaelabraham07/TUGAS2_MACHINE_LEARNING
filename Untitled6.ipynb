IMPORT LIBRARY

```python
# Import library utama
import pandas as pd
import numpy as np

# Visualisasi
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing
from sklearn.preprocessing import PowerTransformer, Normalizer

# Clustering & evaluasi
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Load dataset (pakai separator ; karena format CSV)
df = pd.read_csv('/content/drive/MyDrive/Machine_Learning/TUGAS 2/DataSpotify.csv', sep=';', low_memory=False)

# Lihat 5 data awal
df.head()
```

DATA UNDERSTANDING

```python
# Informasi dataset
df.info()
```

```python
# Statistik deskriptif
df.describe()
```

```python
# Cek missing value
df.isnull().sum()
```

```python
# Ukuran data
df.shape
```

DATA PREPROCESSING

```python
# Hapus kolom yang tidak digunakan
drop_cols = ['id', 'name', 'artist']
df = df.drop(columns=[col for col in drop_cols if col in df.columns])
```

```python
# Hapus missing value
df = df.dropna()
```

```python
# Hapus duplikasi
df = df.drop_duplicates()

print("After cleaning:", df.shape)
```

FILTER DATA

```python
# Filter data untuk mengurangi noise (penting untuk silhouette)
df = df[
    (df['energy'] > 0.4) & (df['energy'] < 0.8)
]

print("After filtering:", df.shape)
```

FEATURE SELECTION

```python
# Pilih fitur yang paling berpengaruh untuk clustering
features = ['energy', 'acousticness']

df = df[features]

# Pastikan semua numeric
df = df.apply(pd.to_numeric, errors='coerce')
df = df.dropna()

print(df.dtypes)
```

REMOVE OUTLIEER

```python
# Menghapus outlier menggunakan metode IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.0 * IQR)) |
          (df > (Q3 + 1.0 * IQR))).any(axis=1)]

print("After outlier removal:", df.shape)
```

EDA (EXPLORATORY DATA ANALYSIS)

```python
# Histogram distribusi data
df.hist(figsize=(12,8))
plt.show()
```

```python
# Korelasi antar fitur
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

FEATURE ENGINEERING

```python
# Transformasi data agar lebih normal
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

DATA SPLITTING

```python
# Gunakan 80% data untuk clustering
persen = 0.8
n_data = int(len(X_scaled) * persen)

X_used = X_scaled[:n_data]

print("Total data:", len(X_scaled))
print("Data digunakan:", len(X_used))
```

SAMPLE DATA

```python
# Ambil sebagian data agar tuning cepat dan stabil
sample_size = 500

if len(X_used) > sample_size:
    idx = np.random.choice(len(X_used), sample_size, replace=False)
    X_sample = X_used[idx]
else:
    X_sample = X_used

print("Data untuk tuning:", len(X_sample))
```

TUNING ALGORITMA K-MEANS

```python
sil_scores = []
k_range = range(2, 10)

for k in k_range:
    model = KMeans(n_clusters=k, random_state=42, n_init=50)
    labels = model.fit_predict(X_sample)

    sil = silhouette_score(X_sample, labels)
    sil_scores.append(sil)

    print(f"k={k}, silhouette={sil}")

# Ambil K terbaik
best_k = k_range[np.argmax(sil_scores)]
best_sil_kmeans = max(sil_scores)
print("Best K:", best_k)
print("Best Silhouette KMeans:", best_sil_kmeans)
```

TUNING ALGORITMA AGGLOMERATIVE



```python
from sklearn.cluster import AgglomerativeClustering

sil_scores_agglo = []
k_range = range(2, 10)

for k in k_range:
    model = AgglomerativeClustering(n_clusters=k)
    labels = model.fit_predict(X_sample)

    sil = silhouette_score(X_sample, labels)
    sil_scores_agglo.append(sil)

    print(f"Agglo k={k} -> {sil}")

best_k_agglo = k_range[np.argmax(sil_scores_agglo)]
best_sil_agglo = max(sil_scores_agglo)

print("\nBest Agglo K:", best_k_agglo)
print("Best Silhouette Agglo:", best_sil_agglo)
```

TUNING ALGORITMA DBSCAN

```python
from sklearn.cluster import DBSCAN

best_eps = 0
best_sil_dbscan = -1

for eps in np.arange(0.1, 0.5, 0.05):
    model = DBSCAN(eps=eps, min_samples=5)
    labels = model.fit_predict(X_sample)

    # Pastikan lebih dari 1 cluster
    if len(set(labels)) > 1:
        sil = silhouette_score(X_sample, labels)
        print(f"DBSCAN eps={eps} -> {sil}")

        if sil > best_sil_dbscan:
            best_sil_dbscan = sil
            best_eps = eps

print("\nBest eps DBSCAN:", best_eps)
print("Best Silhouette DBSCAN:", best_sil_dbscan)
```

MODEL COMPARISON

```python
# =========================
# PERBANDINGAN MODEL
# =========================

comparison = pd.DataFrame({
    'Model': ['KMeans', 'Agglomerative', 'DBSCAN'],
    'Silhouette': [best_sil_kmeans, best_sil_agglo, best_sil_dbscan]
})

comparison = comparison.sort_values(by='Silhouette', ascending=False)
comparison
```

PILIH MODEL

```python
# =========================
# PILIH MODEL TERBAIK
# =========================

best_model_name = comparison.iloc[0]['Model']
print("Best Model:", best_model_name)
```

MODEL FINAL

```python
# =========================
# MODEL FINAL
# =========================

if best_model_name == 'KMeans':
    final_model = KMeans(
        n_clusters=best_k,
        random_state=42,
        n_init=50
    )

elif best_model_name == 'Agglomerative':
    final_model = AgglomerativeClustering(
        n_clusters=best_k_agglo
    )

elif best_model_name == 'DBSCAN':
    final_model = DBSCAN(
        eps=best_eps,
        min_samples=5
    )

labels_final = final_model.fit_predict(X_used)
```

EVALUASI FINAL

```python
# =========================
# EVALUASI MODEL FINAL
# =========================

sil_final = silhouette_score(X_used, labels_final)
dbi_final = davies_bouldin_score(X_used, labels_final)

print("Final Model:", best_model_name)
print("Silhouette:", sil_final)
print("DBI:", dbi_final)
```

HASIL DAN RINGKASAN CLUSTER

```python
# =========================
# HASIL CLUSTER
# =========================

df_cluster = df.iloc[:n_data].copy()
df_cluster['cluster'] = labels_final

df_cluster.head()
```

```python
# =========================
# RINGKASAN CLUSTER
# =========================

cluster_summary = df_cluster.groupby('cluster').mean()
cluster_summary
```

INTERPRETASI DAN VISUALISASI

```python
# =========================
# INTERPRETASI CLUSTER
# =========================

for i, row in cluster_summary.iterrows():
    print(f"\nCluster {i}")

    if row['energy'] > cluster_summary['energy'].mean():
        print("- Lagu energik")
    else:
        print("- Lagu santai")
    if row['acousticness'] > cluster_summary['acousticness'].mean():
        print("- Lagu akustik")
    else:
        print("- Lagu tidak akustik")
```

```python
# =========================
# VISUALISASI
# =========================

plt.scatter(X_used[:,0], X_used[:,1], c=labels_final)
plt.title(f'Clustering ({best_model_name})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
